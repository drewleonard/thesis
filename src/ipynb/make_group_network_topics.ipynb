```python
from __future__ import division
import pandas as pd
import sys
```

```python
df = pd.read_csv('/Users/drewnleonard/Documents/thesis/data/csv/df_interests_spline_age_49.csv')
```

```python
# Remove unavailable
df = df[df["AccountGroup"] != 'Unavailable']
df = df[df["primary_topic"] != 'unavailable']
df = df[['AccountGroup','primary_topic']]
```

```python
group_to_topic = {}
for n, e in df.iterrows():
    
    group = e['AccountGroup']
    topic = e['primary_topic']
    
    if group not in group_to_topic:
        group_to_topic[group] = {}
    
    if topic not in group_to_topic[group]:
        group_to_topic[group][topic] = 0
    
    group_to_topic[group][topic] += 1
```

```python
group_to_group = {}
for group,topic_map in group_to_topic.iteritems():
    for topic,count in topic_map.iteritems():
        
        for inner_group,inner_topic_map in group_to_topic.iteritems():
            for inner_topic,inner_count in inner_topic_map.iteritems():
                
                if group != inner_group and topic == inner_topic:
                    shared = min(count,inner_count)
                    
                    if inner_group not in group_to_group:
                        group_to_group[inner_group] = {}
                    if group not in group_to_group:
                        group_to_group[group] = {}
                    
                    if group not in group_to_group[inner_group]:
                        group_to_group[inner_group][group] = 0
                    if inner_group not in group_to_group[group]:
                        group_to_group[group][inner_group] = 0
                    
                    group_to_group[inner_group][group] += shared
                    group_to_group[group][inner_group] += shared
```

```python
group1l = []
group2l = []
weightl = []

for group1, group1_map in group_to_group.iteritems():
    for group2, weight in group1_map.iteritems():
        
        group1l.append(group1)
        group2l.append(group2)
        weightl.append(weight)
```

```python
group_to_group_df = pd.DataFrame()
group_to_group_df['group1'] = group1l
group_to_group_df['group2'] = group2l
group_to_group_df['weight'] = weightl
```

```python
node_name = []
node_id = []
n = 0
seen = set()
for n, e in df.iterrows():
    group = e['AccountGroup']
    if group not in seen:
        node_name.append(group)
        node_id.append(n)
        n += 1
    seen.add(group)
```

```python
node_df = pd.DataFrame()
node_df['node'] = node_name
node_df['id'] = node_id
```

```python
node_df.to_csv('/Users/drewnleonard/Documents/thesis/data/csv/group_node_topic.csv',index=False)
group_to_group_df.to_csv('/Users/drewnleonard/Documents/thesis/data/csv/group_to_group_topic.csv',index=False)
```
